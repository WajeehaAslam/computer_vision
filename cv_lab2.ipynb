**ellipse2Poly**

```python
import cv2 as cv
from google.colab.patches import cv2_imshow


image = cv.imread('/content/pandas.jpg')


center = (77, 67)
axes = (80, 40)
angle = 0
start_angle = 0
end_angle = 360


pts = cv.ellipse2Poly(center, axes, angle, start_angle, end_angle, delta=75)


cv.polylines(image, [pts], isClosed=True, color=(0, 0, 255), thickness=2)


cv2_imshow(image)
```

**Filling Convex Polygons**

```python
import cv2 as cv
from google.colab.patches import cv2_imshow


image = cv.imread('/content/pandas.jpg')


center = (77, 67)
axes = (80, 40)
angle = 0
start_angle = 0
end_angle = 360


pts = cv.ellipse2Poly(center, axes, angle, start_angle, end_angle, delta=45)

cv.fillConvexPoly(image, pts, (0, 255, 0))

cv2_imshow(image)
```

**Drawing Rectangles**

```python
import cv2 as cv
from google.colab.patches import cv2_imshow


image = cv.imread('/content/pandas.jpg')


pt1 = (50, 50)
pt2 = (100, 100)


cv.rectangle(image, pt1, pt2, (0, 0, 255), 3)


cv2_imshow(image)
```

***Drawing* Lines and Arrowed Lines**

```python
import cv2 as cv
from google.colab.patches import cv2_imshow


image = cv.imread('/content/pandas.jpg')


start_point = (0, 50)
end_point = (200, 50)
cv.line(image, start_point, end_point, (0, 0, 255), thickness=3)  # Red line


start_point2 = (50, 100)
end_point2 = (100, 100)
cv.arrowedLine(image, start_point2, end_point2, (0, 255, 0), thickness=3)  # Green arrow

cv2_imshow(image)
```
