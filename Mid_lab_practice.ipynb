Lab Task 1:
Basic Image Operations
Objective:
Perform basic image operations like resizing, rotating, and flipping.
Instructions:
• Read any open-source image.
• Resize the image to half its original dimensions.
• Rotate the resized image 180 degrees clockwise.
• Flip the rotated image vertically.



```python
import cv2
from google.colab.patches import cv2_imshow
img =cv2.imread('/content/pen.jpg')

cv2_imshow(img)

```

```python
# Step 2: Resize the image to half its original dimensions

height , width = img.shape[:2]
resized_img=cv2.resize(img,(width//2,height//2))
cv2_imshow(resized_img)
```

```python
# Step 3: Rotate the resized image 180 degrees clockwise

rotated_img = cv2.rotate(resized_img,cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated_img)
```

```python
# Step 4: Flip the rotated image vertically
flipped= cv2.flip(img, 0)
cv2_imshow(flipped)  #original image flipped
print("original image flipped")

flipped_img = cv2.flip(rotated_img,0)
cv2_imshow(flipped_img)  #rotated image flip
print("rotated image flip")
```

Lab Task 2:
Image Filtering Techniques
Objective:
Apply image filtering techniques, including blurring and grayscale conversion.
Instructions:
• Take the image from Exercise 2.
• Blur the image using a Gaussian blur with a kernel size of 5x5.
• Convert the blurred image to grayscale.
• Saved the image

```python
# Step 2: Apply Gaussian Blur with kernel size 9x9
blurred_img = cv2.GaussianBlur(img,(9,9),0)
cv2_imshow(blurred_img)
```

```python
# Step 3: Convert the blurred image to grayscale
gray_img=cv2.cvtColor(blurred_img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
# Saved the image
cv2.imwrite('blurred_grayscale_image.jpg', gray_img)
```

Converting Ellipses to Polygons

```python
img = cv2.imread('/content/pen.jpg')

center = (100,100)
axes = (100,50)
angle = 45
start_angle = 0
end_angle = 360


ellips2Poly = cv2.ellipse2Poly(center, axes, angle, start_angle, end_angle, delta=45) # Added delta for point density

poly =cv2.polylines(img, [ellips2Poly], isClosed=True, color=(0, 0, 255), thickness=2)

cv2_imshow(poly)
```

Filling Convex Polygons

```python
img = cv2.imread('/content/pen.jpg')

center = (100,100)
axes = (100,50)
angle = 45
start_angle = 0
end_angle = 360


ellips2Poly = cv2.ellipse2Poly(center, axes, angle, start_angle, end_angle, delta=45) # Added delta for point density

poly =cv2.fillConvexPoly(img, ellips2Poly,  color=(55, 99, 66))

cv2_imshow(poly)
```

```python
path = "/content/Animals"
os.chdir(path)

mean_height =0
mean_width =0
```

```python
no_of_imges = len([file for file in os.listdir('.') if file.endswith((".jpg","jpeg","png"))])
print("number of images",no_of_imges)
```

```python
for file in os.listdir('.'):
  if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith("png"):
    im = Image.open(os.path.join(path,file))
    width,height =im.size
    mean_width+=width
    mean_height+=height
```

```python
mean_height=int(mean_height/no_of_imges)
mean_width=int(mean_width/no_of_imges)
```

```python
for file in os.listdir('.'):
  if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith("png"):
    im = Image.open(os.path.join(path,file))
    im_resize = im.resize((mean_width,mean_height),Image.LANCZOS)
    im_resize.save(file,'JPEG',quality=95)
    print(f"{file} is resized")
```
