```python
import cv2
from google.colab.patches import cv2_imshow

# Read the original image
img = cv2.imread("/content/bunny.jpg")

# Display original image
cv2_imshow(img)
cv2.waitKey(0)

# Convert to graycsale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(img_gray, (3,3), 0)

# Sobel Edge Detection
sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
# Display Sobel Edge Detection Images
cv2_imshow(sobelx)
cv2.waitKey(0)
cv2_imshow(sobely)
cv2.waitKey(0)
cv2_imshow(sobelxy)
cv2.waitKey(0)

# Canny Edge Detection
edges = cv2.Canny(image=img_blur, threshold1=100, threshold2=200) # Canny Edge Detection
# Display Canny Edge Detection Image
cv2_imshow(edges)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
import cv2
from google.colab.patches import cv2_imshow


img = cv2.imread("/content/bunny.jpg")
cv2_imshow(img)
cv2.waitKey(0)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
img_blur = cv2.GaussianBlur(img_gray, (3, 3), 0)
cv2_imshow(img_blur)
```

```python
sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
cv2_imshow(sobelx)
cv2.waitKey(0)
```

```python
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)
cv2_imshow(sobely)
cv2.waitKey(0)
```

```python
sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)
cv2_imshow(sobelxy)
cv2.waitKey(0)
```

```python
#Canny Edge Detection ---
edges = cv2.Canny(image=img_blur, threshold1=100, threshold2=200)
cv2_imshow(edges)
cv2.waitKey(0)
```

```python
#Laplacian Edge Detection
laplacian = cv2.Laplacian(img_blur, cv2.CV_64F)
cv2_imshow(laplacian)
cv2.waitKey(0)
```

```python
#Scharr Edge Detection ---
scharrx = cv2.Scharr(img_blur, cv2.CV_64F, 1, 0)
scharry = cv2.Scharr(img_blur, cv2.CV_64F, 0, 1)
scharr_combined = cv2.addWeighted(scharrx, 0.5, scharry, 0.5, 0)
cv2_imshow(scharr_combined)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read and display original image
img = cv2.imread("/content/bunny.jpg")
cv2_imshow(img)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# ---- Custom Blurring Filter using filter2D ----
kernel = np.ones((5,5), np.float32) / 25   # simple averaging kernel
custom_blur = cv2.filter2D(gray, -1, kernel)
cv2_imshow(custom_blur)

# ---- Another custom kernel (weighted blur) ----
kernel2 = np.array([[1,2,1],
                    [2,4,2],
                    [1,2,1]], np.float32)
kernel2 = kernel2 / kernel2.sum()
weighted_blur = cv2.filter2D(gray, -1, kernel2)
cv2_imshow(weighted_blur)

# ---- Compare with standard OpenCV blurs ----
blur = cv2.blur(gray, (5,5))
gauss = cv2.GaussianBlur(gray, (5,5), 0)
median = cv2.medianBlur(gray, 5)

cv2_imshow(blur)
cv2_imshow(gauss)
cv2_imshow(median)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
